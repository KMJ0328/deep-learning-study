```python
import tensorflow as tf

# 배달 시간은 거리에 비례한다는 가정하에 간단하게 수식을 만들어 보자.
# 거리에 가중치를 곱하고 bias(편향, y좌표 위치)를 더한 식이다.
# 가중치의 변화는 그래프의 기울기 변화를 뜻하며 편향은 y좌표의 위치를 의미한다.
# 배달 시간 = 거리 * w + b
# 33분 = 850m * w + b

label = 33
distance = 850

# w와 b는 변수로 사용할 것이므로 tf.Variable로 지정
w = tf.Variable(0.)
b = tf.Variable(0.)

# 변수만 출력할 경우 여러 정보들을 볼 수 있음.
print(w)
# 수치만 출력할 경우 numpy 함수를 이용할 수 있음.
print(w.numpy())
```

```python
# 손실함수는 예측값과 실제값의 차이를 줄여 알고리즘의 성능을 향상시킬 때 사용하는 함수이다.
def loss_func() :
  # 예측값을 얻기 위한 수식
  pred = distance * w + b

  # 실제값과 예측값의 차액이 음수가 될 수 있음으로 square 함수 처리
  return tf.square(label - pred)
```

```python
# 최적화 함수 중 평균적으로 성능이 좋음 Adam 함수 사용
# learning_rate는 0.1을 지정했으며 0.01, 0.001 등을 사용할 수 있다.
opt = tf.keras.optimizers.Adam(learning_rate = 0.1)

for i in range(1000) :
  # 손실함수가 최소값을 갖기 위한 w, b 구하기
  opt.minimize(loss_func, var_list=[w, b])
  print(w.numpy(), b.numpy())
```

```python
distance * w + b
```

# Quiz
> 위의 코드를 이용하여 손실함수가 최소값을 유지할 때의 최소 반복 횟수를 구하시오.

```python
opt = tf.keras.optimizers.Adam(learning_rate = 0.1)

w_back = tf.Variable(0.)
b_back = tf.Variable(0.)

cnt = tf.Variable(0.)

while True :
  opt.minimize(loss_func, var_list=[w, b])
  if (w.numpy() == w_back.numpy()) and (b.numpy() == b_back.numpy()) :
    break
  # 참조가 같아져서 항상 같은 값이 나타남.
  # w_back = w

  w_back = tf.Variable(w.numpy())
  b_back = tf.Variable(b.numpy())

  cnt.assign_add(1)

print(cnt.numpy())
```
